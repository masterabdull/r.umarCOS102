```python
import numpy as np
import cmath

def solve_quadratic(a, b, c):
    """Find roots of a quadratic equation: ax^2 + bx + c = 0"""
    discriminant = (b**2) - (4*a*c)
    
    if discriminant > 0:
        root1 = (-b + cmath.sqrt(discriminant)) / (2*a)
        root2 = (-b - cmath.sqrt(discriminant)) / (2*a)
        return [root1, root2]
    elif discriminant == 0:
        root = -b / (2*a)
        return [root]
    else:
        root1 = (-b + cmath.sqrt(discriminant)) / (2*a)
        root2 = (-b - cmath.sqrt(discriminant)) / (2*a)
        return [root1, root2]

def solve_polynomial(coefficients):
    """Find roots of a cubic or quartic equation"""
    return np.roots(coefficients)

def main():
    print("Equation Solver: Quadratic, Cubic, or Quartic")
    choice = input("Choose the type of equation to solve (quadratic/cubic/quartic): ").strip().lower()

    if choice == "quadratic":
        a = float(input("Enter coefficient a: "))
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter coefficient c: "))
        roots = solve_quadratic(a, b, c)

    elif choice == "cubic":
        a = float(input("Enter coefficient A: "))
        b = float(input("Enter coefficient B: "))
        c = float(input("Enter coefficient C: "))
        d = float(input("Enter coefficient D: "))
        roots = solve_polynomial([a, b, c, d])

    elif choice == "quartic":
        a = float(input("Enter coefficient A: "))
        b = float(input("Enter coefficient B: "))
        c = float(input("Enter coefficient C: "))
        d = float(input("Enter coefficient D: "))
        e = float(input("Enter coefficient E: "))
        roots = solve_polynomial([a, b, c, d, e])

    else:
        print("Invalid choice! Please enter quadratic, cubic, or quartic.")
        return

    print("\nRoots of the equation:")
    for root in roots:
        print(root)

if __name__ == "__main__":
    main()
```
